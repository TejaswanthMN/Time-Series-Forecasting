```python
# importing necessary libraries

import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.metrics import mean_squared_error
```

```python
# importing the data

df = pd.read_csv('drive/MyDrive/Data/daily-min-temperatures.csv', parse_dates = [0])
df.head()
```

```python
df['t'] = df['Temp'].shift(1)
df.head()
```

```python
# Train test split where we are taking last 7 values as the test data

train, test = df[1: df.shape[0] - 7], df[df.shape[0] - 7:]
```

```python
train.head()
```

```python
test.head()
```

```python
train_X, train_Y = train['t'], train['Temp']
test_X, test_Y = test['t'], test['Temp']
```

```python
predictions = test_X.copy()
```

```python
# Calculating the error

mse = mean_squared_error(test_Y, predictions)
mse
```

```python
# Plotting the actual and predicted values

plt.plot(test_Y)
plt.plot(predictions, color = 'red')
```
